#### Inverted dropout regularization

```python
import numpy as np
```

```python
a3 = np.random.rand(4, 5)
print("Activation a3 =\n", a3)
print("----------------------------------------------------------")
keep_prob = 0.8
d3 = np.random.rand(a3.shape[0], a3.shape[1]) < keep_prob
a3 = a3 * d3
print("After applying dropout mask a3 =\n", a3)
print("----------------------------------------------------------")
a3 = a3 / keep_prob
print("After scaling a3 =\n", a3)
```
